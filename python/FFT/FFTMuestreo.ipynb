# FFT con Python. (Diferente muestreo la señal)

### Importando los paquetes necesarios

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%pylab inline --no-import-all
```

### Muesteando la señal coseno con 500 valores.

```python
t = np.linspace(0, 2*np.pi, 500, endpoint=True)
f = 1.0/1.0 # Frequency in Hz
A = 1.0 # Amplitude in Unit
s = A * np.cos(2*np.pi*f*t) # Signal
```

### Graficando la señal muestreada obtenida

```python
plt.plot(t,s)
plt.xlabel('Time ($s$)')
plt.ylabel('Amplitude ($Unit$)')
```

### Haciendo la transformada de fourier discreta

```python
Y = np.fft.fft(s)
```

### Graficando el resultado. Tenemos:

```python
plt.plot(Y)
```

## Cambiemos ahora el valor de la frecuencia de muestreo con solo 25 valores y veamos que sucede:

```python
t = np.linspace(0, 2*np.pi, 25, endpoint=True)
f = 1.0/1.0 # Frequency in Hz
A = 1.0 # Amplitude in Unit
s = A * np.cos(2*np.pi*f*t) # Signal
```

## Graficando la nueva señal coseno muestrada:

```python
plt.plot(t,s)
plt.xlabel('Time ($s$)')
plt.ylabel('Amplitude ($Unit$)')
```

## Haciendo la FFT

```python
Y = np.fft.fft(s)
```

## Ploteando

```python
plt.plot(Y)
```

# Como observamos hay un solapamiento de la señal en el resultado. Si hacemos más pequeño el valor de la muestra se observa que la señal llega a convertirse en un solo pulso.
